## Calculate correlations

```python
def calculate_correlations(cost_matrix, p1, p2):
    import scipy as sp
    pseud_dist = []
    for x in p1:
        pdt = []
        for y in p2:
            pdt.append(abs(x-y))
        pseud_dist.append(pdt)

    correlations = []
    for i, el in enumerate(cost_matrix):
        corr = sp.stats.pearsonr(cost_matrix[i], pseud_dist[i])[0]
        correlations.append(corr)
    return correlations
```

```python
def calculate_single_correlation(cost_matrix, dist_cells1, dist_cells2):
    distances = []
    for x in dist_cells1:
        for y in dist_cells2:
            distances.append(abs(x-y))
    costs = []
    for i in range(len(cost_matrix)):
        for j in range(len(cost_matrix[i])):
            costs.append(cost_matrix[i][j])

    import scipy as sp
    corr = sp.stats.pearsonr(costs, distances)[0]
    return corr
```

```python
dataset1 = anndata.read("d1.h5ad")
dataset2 = anndata.read("d2.h5ad")

config = {
    "compare_lineages_pseudocells": False,
    "compare_lineages_pseudocells_label": "pseudocells_100",
    "compare_trajectories_pseudocells": False,
    "compare_trajectories_pseudocells_label": "pseudocells_100",
    "align_pseudocells": False,
}

matching_lineages, matching = dynchro.tl.get_matching_lineages(
    [dataset1, dataset2],
    config=config,
)

result = dynchro.tl.align_trajectories(
    matching,
    pseudocells=False,
)

# get pseudotime differences
# get differences in cost matrix
cost_matrix = result[0].uns["alignment_costs_dataset1_lineage1_dataset2_lineage1"]
pseudotime_d1_l1 = result[0][result[0].obs.lineage1, :].obs.pseudotime
pseudotime_d1_l1 = pseudotime_d1_l1[np.argsort(pseudotime_d1_l1)]
pseudotime_d2_l1 = result[1][result[1].obs.lineage1, :].obs.pseudotime
pseudotime_d2_l1 = pseudotime_d2_l1[np.argsort(pseudotime_d2_l1)]

correlations_non_shuffled = calculate_correlations(cost_matrix, pseudotime_d1_l1, pseudotime_d2_l1)

import matplotlib.pyplot as plt
plt.scatter(range(len(correlations_non_shuffled)), correlations_non_shuffled)
```

```python
all_correlations = []
single_correlations = []

for i in range(100):

    result0 = anndata.read(f"shuffled/d1_shuffled_{i}_aligned0.h5ad")
    result1 = anndata.read(f"shuffled/d1_shuffled_{i}_aligned1.h5ad")

    match = [x for x in result0.uns.keys() if x.startswith("alignment_costs")][0]
    parts = match.split("_")
    lin1 = parts[3]
    lin2 = parts[5]

    lin1id = int(lin1[-1])
    lin2id = int(lin2[-1])

    
    # get pseudotime differences
    # get differences in cost matrix
    cost_matrix = result0.uns[match]
    pseudotime_d1_l1 = result0[result0.obs[f"lineage{lin1id}"], :].obs.pseudotime
    pseudotime_d2_l1 = result1[result1.obs[f"lineage{lin2id}"], :].obs.pseudotime
    pseudotime_d1_l1 = pseudotime_d1_l1[np.argsort(pseudotime_d1_l1)]
    pseudotime_d2_l1 = pseudotime_d2_l1[np.argsort(pseudotime_d2_l1)]

    correlations = calculate_correlations(cost_matrix, pseudotime_d1_l1, pseudotime_d2_l1)
    single_correlations.append(calculate_single_correlation(cost_matrix, pseudotime_d1_l1, pseudotime_d2_l1))

    all_correlations.append(correlations)

    import matplotlib.pyplot as plt
    plt.ylim = (-1, 1)
    plt.scatter(range(len(correlations)), correlations)
```
